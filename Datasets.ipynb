```python
from sklearn.datasets import make_blobs, make_classification, make_moons, make_regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
blob = make_blobs(n_samples=200, n_features = 2, random_state = 42)
```

```python
blob_df = pd.DataFrame(blob[0], blob[1])
blob_df = blob_df.reset_index(len(blob_df))
blob_df['cat'] = blob_df['index']
blob_df = blob_df.drop('index', axis =1)
blob_df.head()
```

```python
blob_df.to_pickle('./data/Blob_data')
```

```python
cl = make_classification(n_samples = 500, n_features= 10)
```

```python
cl_df = pd.DataFrame(cl[0], cl[1])
```

```python
cl_df.head()
```

```python
cl_df.to_pickle('./data/Classification_data')
```
